```python
import os
import pandas as pd
import requests
import re
from tqdm import tqdm
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer, util
#from specialty_keywords import specialty_keywords  # Your weighted keyword dictionary

# ========== SETTINGS ==========
CSV_PATH = "FINAL_dataset.csv"  
DOWNLOAD_DIR = "downloads"
OUTPUT_FILE = "FINAL_dataset_clinical_specialty.csv"
FAILED_URL_LOG = "failed_urls.txt"

# ========== SETUP ==========
os.makedirs(DOWNLOAD_DIR, exist_ok=True)

semantic_model = SentenceTransformer('all-MiniLM-L6-v2')
specialties = list(specialty_keywords.keys())
specialty_phrases = [' '.join(s.split('_')) for s in specialties]
specialty_embeddings = semantic_model.encode(specialty_phrases, convert_to_tensor=True)

# ========== FUNCTIONS ==========

def weighted_keyword_score(text):
    import re
    text = text.lower()
    scores = {}
    for specialty, keywords in specialty_keywords.items():
        score = 0
        for keyword, weight in keywords.items():
            # Use word boundaries for whole word match
            pattern = r'\b' + re.escape(keyword.lower()) + r'\b'
            matches = re.findall(pattern, text)
            score += len(matches) * weight
        scores[specialty] = score
    return scores

def semantic_similarity(text):
    text_embedding = semantic_model.encode(text[:1000], convert_to_tensor=True)  # can Truncate for efficiency
    similarities = util.cos_sim(text_embedding, specialty_embeddings)[0].cpu().tolist()
    best_index = similarities.index(max(similarities))
    return specialties[best_index], similarities[best_index]

def extract_text_from_pdf(filepath, max_pages=500):
    try:
        reader = PdfReader(filepath)
        text = ''.join((page.extract_text() or '') for page in reader.pages[:max_pages])
        return text.strip()
    except Exception as e:
        return ''

SEMANTIC_THRESHOLD = 0.35

def classify_document(text):
    keyword_scores = weighted_keyword_score(text)
    threshold = 12  # Customizable threshold

    # Find all specialties with score >= threshold
    matched_specialties = [spec for spec, score in keyword_scores.items() if score >= threshold]

    if matched_specialties:
        return matched_specialties, keyword_scores, "keywords"
    
    # Fallback to semantic
    semantic_label, sim_score = semantic_similarity(text)
    if sim_score < SEMANTIC_THRESHOLD:
        return ["unclassified"], keyword_scores, f"semantic (low score: {sim_score:.2f})"
    return [semantic_label], keyword_scores, f"semantic (score: {sim_score:.2f})"

def safe_filename_from_url(url):
    return os.path.join(DOWNLOAD_DIR, re.sub(r'[^\w\-_.]', '_', url.split('/')[-1]))

def download_pdf(url, filename):
    try: 
        headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
                }

        r = requests.get(url, headers=headers, timeout=300)
        if r.status_code == 200:
            with open(filename, 'wb') as f:
                f.write(r.content)
            return True
        else:
            return False
    except Exception:
        return False

def process_pdfs(csv_file):
    df = pd.read_csv(csv_file)
    urls = df['URL'].dropna().tolist()
    results = []
    failed_urls = []

    for url in tqdm(urls, desc="Processing URLs"):
        filename = safe_filename_from_url(url)

        if not os.path.exists(filename):
            success = download_pdf(url, filename)
            if not success:
                failed_urls.append(url)
                results.append({
                    "url": url,
                    "predicted_specialty": "error",
                    "classification_method": "Download failed (404 or other issue)"
                })
                continue

        text = extract_text_from_pdf(filename)
        if not text:
            failed_urls.append(url)
            results.append({
                "url": url,
                "predicted_specialty": "error",
                "classification_method": "Text extraction failed"
            })
            continue

        label, scores, method = classify_document(text)
        results.append({
    "url": url,
    "predicted_specialty": ', '.join(label) if isinstance(label, list) else label,
    "classification_method": method,
    **scores
})

    # Save results and failed URLs
    pd.DataFrame(results).to_csv(OUTPUT_FILE, index=False)
    if failed_urls:
        with open(FAILED_URL_LOG, "w") as f:
            f.write('\n'.join(failed_urls))

    print(f"\n✅ Finished processing.\n- Results: {OUTPUT_FILE}\n- Failed URLs: {FAILED_URL_LOG if failed_urls else 'None'}")

# ========== MAIN ==========
if __name__ == "__main__":
    process_pdfs(CSV_PATH)
```
